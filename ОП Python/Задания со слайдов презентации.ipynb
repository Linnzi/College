### Задание D

```python
class Parrot:

    def __init__(self, phrase): # Принимает начальную фразу для попугая
        self.phrase = phrase
    
    def say(self, repeat = 1):
        print((self.phrase + " ") * repeat)
    
    def newText(self, new_phrase): # Изменение фразы попугаю
        self.phrase = new_phrase

p = Parrot("Гав!")
p.say()
p.newText("Мяу!")
p.say(3)
```

### Задание E

```python
class Parrot:
    
    def __init__(self, phrase): # Принимает начальную фразу для попугая
        self.phrase = [phrase] # Список фраз
    
    def say(self, repeat = 1):
        phrase = rn.choice(self.phrase) # Используется random обозначеный как rn в самом начале работы
        print(" ".join([phrase] * repeat))
    
    def learn(self, phrase):
        self.phrase.append(phrase) # Добавляем в список фраз попугая новую фразу

p = Parrot("Гав!")
p.say()
p.learn("Мяу!")
p.say()
p.say(3) # Выводит 3 раза подряд Мяу! или Гав!
```

### Задание A

```python
class lamprow:

    def __init__(self):
        self.__state = "00000000"

    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, value):
        if len(value) == 8:
            self.__state = value
        else:
            self.__state = "00000000"

    def show(self):
        morzh = ""
        for i in self.__state:
            if i == "0":
                morzh += "-"
            else:
                morzh += "*"
        print(morzh)

lamps = lamprow()
lamps.show()
lamps.state = "10101010"
print(lamps.state)
lamps.show()
```

### Задание B

```python
class lamprow:

    def __init__(self, count = 8):
        self.count = count
        self.__state = "0" * count

    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, value):
        if len(value) == self.count:
            self.__state = value
        else:
            print("Ошибка")
            self.__state = "0" * self.count

    def show(self):
        morzh = ""
        for i in self.__state:
            if i == "0":
                morzh += "-"
            else:
                morzh += "*"
        print(morzh)

lamps = lamprow(6)
lamps.show()
lamps.state = "101010"
print(lamps.state)
lamps.show()
lamps.state = "10101010"
print(lamps.state)
lamps.show()
```

### Задание C

```python
class lamprow:

    def __init__(self, count = 8):
        self.count = count
        self.__state = "0" * count

    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, value):
        if len(value) == self.count and all(ch in "012" for ch in value):
            self.__state = value
        else:
            print("Ошибка")
            self.__state = "0" * self.count

    def show(self):
        morzh = ""
        for i in self.__state:
            if i == "0":
                morzh += "-"
            elif i == "1":
                morzh += "*"
            else:
                morzh += "o"
        print(morzh)

lamps = lamprow(6)
lamps.show()
lamps.state = "102102"
print( lamps.state )
lamps.show()
lamps.state = "10201010"
print( lamps.state )
lamps.show()
```

### Задание D

```python
class lamprow:
    def __init__(self, count = 8):
        self.count = count
        self.__state_int = 0

    @property
    def state(self):
        result = ""
        temp = self.__state_int
        for i in range(self.count):
            digit = temp % 10
            result = str(digit) + result
            temp //= 10
        return result

    @state.setter
    def state(self, value):
        if len(value) == self.count and all(i in "012" for i in value):
            self.__state_int = 0
            for i in value:
                self.__state_int = self.__state_int * 10 + int(i)
        else:
            print("Ошибка")
            self.__state_int = 0

    def show(self):
        morzh = ""
        temp = self.__state_int
        for i in range(self.count):
            digit = temp % 10
            if digit == 0:
                morzh = "-" + morzh
            elif digit == 1:
                morzh = "*" + morzh
            else:
                morzh = "o" + morzh
            temp //= 10
        print(morzh)

lamps = lamprow(6)
lamps.show()
lamps.state = "102102"
print(lamps.state)
lamps.show()
lamps.state = "10201010"
print(lamps.state)
lamps.show()
```

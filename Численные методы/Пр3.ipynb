```python
import math

x = 1.5  # начальное значение (визуально)
eps = 0.001

while True:
    x1 = math.acos(0.008 * x**3)  # итерационная формула x = arccos(0.008*x^3)
    if abs(x1 - x) < eps:
        break
    x = x1

print(f'Корень: {x:.4f}')
print(f'Проверка: (0.2*{x:.4f})^3 = {(0.2 * x)**3:.4f}, cos({x:.4f}) = {math.cos(x):.4f}')
```

### Задание 1

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,3,500)
y = 0.008*x**3 - np.cos(x)
plt.plot(x,0.008*x**3, x,np.cos(x))
plt.grid(); plt.show()

for i in range(len(x)-1):
    if y[i]*y[i+1]<0: print(f'Корень на [{x[i]:.2f}, {x[i+1]:.2f}]')
```

### Задание 2

```python
import math
f = lambda x: 0.008*x**3 - math.cos(x)
a,b,eps = 1,2,0.001

print(f"{'n':^2} {'x':^20} {'|x_n-x_n-1|':^10}")
n,x_prev=1,a
while True:
    x = a - f(a)*(b-a)/(f(b)-f(a))
    print(f"{n:3.3f} {x:10.4f} {abs(x-x_prev):14.4f}")
    if abs(x-x_prev)<eps: break
    if f(a)*f(x)<0: b=x
    else: a=x
    x_prev,x,n=x,n+1,x_prev
print(f"\nКорень: {x:.4f}")
```

### Задание 3

```python
import math
f = lambda x: 0.008*x**3 - math.cos(x)
f1 = lambda x: 0.024*x**2 + math.sin(x)
x,eps = 1.5,0.001

print(f"{'n':^4} {'x':^12} {'|x_n-x_n-1|':^13}")
n=1
while True:
    x_new = x - f(x)/f1(x)
    print(f"{n:2} {x_new:11.4f} {abs(x_new-x):12.4f}")
    if abs(x_new-x)<eps: break
    x,n=x_new,n+1
print(f"\nКорень: {x_new:.4f}")
```

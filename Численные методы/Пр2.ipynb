## Практическая работа №2

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Задание 1

```python
x = np.linspace(-5, 5, 500)
y1 = 0.008*x**3
y2 = np.cos(x)
y = y1 - y2  # Поиск пересечений

plt.plot(x, y1, label='0.008x³')
plt.plot(x, y2, label='cos(x)')
plt.grid()
plt.show()

print("Интервалы с корнями")
for i in range(len(x)-1):
    if y[i] * y[i+1] < 0:
        print(f'Корень находится на [{x[i]:.2f}, {x[i+1]:.2f}]')
```

### Задание 2

```python
f = lambda x: 0.008*x**3 - np.cos(x)
a, b, eps = 0, 2, 1e-3

print("Метод половинного деления")
print(f"{'n':^3} {'a':^8} {'b':^8} {'x':^8} {'F(x)':^8}")
n = 1
while b - a > 2*eps:
    x = (a+b)/2
    print(f"{n:3} {a:8.4f} {b:8.4f} {x:8.4f} {f(x):8.4f}")
    if f(a)*f(x) < 0: 
        b = x
    else: 
        a = x
    n += 1

print(f"\nКорень: {(a+b)/2:.3f}")
```
